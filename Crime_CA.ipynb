```python
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data with violent crime reports for cities over 100,000 hab from AL to CA
#Data Source: https://ucr.fbi.gov/crime-in-the-u.s/2017/preliminary-report/tables/table-4/state-cuts/alabama-through-california.xls
#loading data 
#indir='example1_crimeCA/'
infile='fbi_ratecrime.xls'
dat_tot=pd.read_excel(infile,header=5)
```

```python
#Displaying header of data
dat_tot
```

```python
#some cleaning
#dat_tot.shape[0]
dat_ca=dat_tot.iloc[22:167,1:5] #only selecting those rows that belong to CA
dat_ca=dat_ca[dat_ca['City'].notnull()].reset_index(drop=True) #Droping values with null 
dat_ca.columns #to see columns names
dat_ca=dat_ca.rename(columns = {'Violent\ncrime':'Violent crime'}) #renaming a column
dat_ca['Crime rate']=dat_ca['Violent crime']/dat_ca['Population1']*100000 #creating crime rate 
```
